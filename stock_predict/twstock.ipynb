```python
import csv
title=['date','capacity','turnover','open','high','low','close','change','transaction']
data=[]
for stock in stock202301:
    strdate= stock.date.strftime("%Y-%m-%d")
    li =strdate,stock.capacity,stock.turnover,stock.open,stock.high,stock.low,stock.close,stock.change,stock.transaction
    data.append(li)

output_data=open('2317_2023_1_to_2023_10_13.csv','w',newline='')
outputwriter = csv.writer(output_data)
outputwriter.writerow(title)

for dataline in data:
    outputwriter.writerow(dataline)
output_data.close()
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = "mingliu"
plt.rcParams["axes.unicode_minus"] = False

import pandas as pd

pdstock = pd.read_csv('2317_2023_1_to_2023_10_13.csv')
pdstock['date']= pd.to_datetime(pdstock['date'])
pdstock.plot(kind='line',figsize=(12,6),x='date',y=['high','close','low'])
```

```python

from plotly.graph_objs import Scatter,Layout
from plotly.offline import plot

data_plotly=[Scatter(x=pdstock['date'],y=pdstock['close'],name='close'),
             Scatter(x=pdstock['date'],y=pdstock['high'],name='high'),
             Scatter(x=pdstock['date'],y=pdstock['low'],name='low')
             ]
plot({"data":data_plotly},auto_open=True)
```
