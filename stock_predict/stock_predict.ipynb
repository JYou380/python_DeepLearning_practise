```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
pdstock = pd.read_csv('2317_2023_1_to_2023_10_13.csv')
close_all = pd.DataFrame(pdstock['close'])
close_all= np.array(close_all).astype('float')
```

```python
#print (close_all)
scalar = MinMaxScaler()
close_all = scalar.fit_transform(close_all)
```

```python
TIME_STEPS = 14

data = []

for i in range (len(close_all)- TIME_STEPS):
    data.append(close_all[i:i+TIME_STEPS+1])
```

```python
print (data)
reshape_data= np.array(data).astype('float64')
```

```python
x = reshape_data[:,:-1]
y = reshape_data[:,-1]
print (y)
```

```python
split = 0.8
split_boundary = int(reshape_data.shape[0] * split)

train_x = x[: split_boundary]
test_x = x[split_boundary:]
train_y = y[: split_boundary]
test_y = y[split_boundary :]
```

```python
print (np.shape(train_x))
print (np.shape(train_y))
```

```python
from keras.models import Sequential
from keras.layers import SimpleRNN,Dropout,Dense
```

```python
model = Sequential()
```

```python
INPUT_SIZE=1
model.add(SimpleRNN(input_shape=(TIME_STEPS,INPUT_SIZE),units=256,unroll=False))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(loss='mse',optimizer="adam",metrics=['accuracy'])
```

```python
model.fit(train_x,train_y,batch_size=100,epochs=100,validation_split=0.1,verbose=2)
```

```python
predict = model.predict(test_x)
```

```python
predict = np.reshape(predict, (predict.size,))
predict_y  = scalar.inverse_transform([[i] for i in predict])
test_y = scalar.inverse_transform(test_y)

print (predict_y[0:10])
print (test_y[0:10])
```

```python
plt.plot(predict_y,'r')
plt.plot(test_y,'k')
```
